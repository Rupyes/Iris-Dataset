```python
data['class']=data['class'].map({'Iris-setosa': 0,
                  'Iris-versicolor': 1,
                  'Iris-virginica':2})
```

```python
data.head()
```

```python
X = data.drop(columns='class', axis=1)
y = data['class'].values
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
```

```python
y_train
```

```python
y_test
```

```python
counts = {'zeros':0, 'ones':0, 'twos':0}
print(counts)
for x in y_train:
    if x == 0:
        counts['zeros'] += 1
    elif x == 1:
        counts['ones'] += 1
    else:
        counts['twos'] += 1

print(counts)
```

```python
counts = {'zeros':0, 'ones':0, 'twos':0}
print(counts)
for x in y_test:
    if x == 0:
        counts['zeros'] += 1
    elif x == 1:
        counts['ones'] += 1
    else:
        counts['twos'] += 1

print(counts)
```

```python
pd.plotting.scatter_matrix(X_train, figsize=(15, 10))
plt.show()
```

```python
pd.plotting.scatter_matrix(X_test, figsize=(15, 10))
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
models = {
    ('Linear', LogisticRegression()),
    ('SVM', SVC()),
    ('CART', DecisionTreeClassifier()),
    ('Ensemble', RandomForestClassifier()),
    ('KNN', KNeighborsClassifier())
}
```

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
names = []
results = []
```

```python
for name, model in models:
    kfold = KFold(n_splits=10, random_state=0)
    result = cross_val_score(model, X, y, scoring='accuracy', cv=kfold)
    names.append(name)
    results.append(result)
    print("{}: {}({})".format(name, result.mean(), result.std()))
```

### confusion matrix

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score
```

```python
# logistic regression

logistic = LogisticRegression()
logistic.fit(X_train, y_train)
y_pred_logistic = logistic.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_logistic)
```

```python
confusion_matrix(y_test, y_pred_logistic)
```

```python
y_test
```

```python
y_pred_logistic
```

```python
# Decision tree
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
y_pred_tree = tree.predict(X_test)
accuracy_score(y_test, y_pred_tree)
```

```python
confusion_matrix(y_test, y_pred_tree)
```

#### classification report of Decision tree

```python
print(classification_report(y_test, y_pred_tree))
```

#### classification report of logistic regression

```python
print(classification_report(y_test, y_pred_logistic))
```

### Support Vector Machine

```python
svm_clf = SVC()
svm_clf.fit(X_train,y_train)
y_pred_svm = svm_clf.predict(X_test)
accuracy_score(y_test, y_pred_svm)
```

```python
confusion_matrix(y_test, y_pred_svm)
```

```python
print(classification_report(y_test, y_pred_svm))
```
